## **Question 1:** *"A signal $x(t)$ is composed by two sines of frequencies $f_1=50$ Hz and $f_2=70$ Hz. What is the minimum sampling frequency according to Nyquist?"*

**Authors' explanation**: This question evaluates the use of precise notation for the Nyquist-Shannon sampling theorem. Since the signal consists of two cosines with different frequencies and contains a discrete frequency component at $f_{max}$ (70Hz), sampling exactly at the Nyquist rate ($F_s = 2f_{max}$) **does not guarantee** perfect reconstruction. Therefore, the sampling condition must be defined by the strict inequality $F_s > 2f_{max}$. Conversely, for signals with infinite frequency components that lack a discrete component at $f_{max}$, the non-strict condition $F_s \ge 2f_{max}$ is sufficient, which is not the case in this query.

For example, if we consider the signal $x(t) = cos(2\pi \times 2.5t - 0.5\pi)$, which has a frequency $f_{max} = 2.5$ Hz, but with a time delay created by the phase $0.5\pi$, and use a $F_s$ of $5$Hz, all the sampling instants will coincide with the zeros of x(t), leading the eventual reconstruction of the signal to fail. The use of $F_s = 2f_{max}$ can eventually also lead to ambiguity between different signals that, when sampled without respecting the strict inequality, cannot be distinguished.

## **Question 2:** *"Can I choose the sampling frequency arbitrarily when I’m using the bilinear transform for designing a digital filter?"*

The role of the sampling frequency $F_s$ in the bilinear transform depends heavily on the intended application. These applications generally fall into three distinct categories:

- **Mimicking $G(s)$**: When the goal is to digitize an existing continuous-time system $G(s)$ so that $H(z)$ closely approximates its overall time-domain behavior, $F_s$ must be chosen based on the bandwidth of the involved signals. This practical selection directly dictates the hardware requirements for the resulting digital control system.
- **Matching a Single Frequency**: If the objective is to ensure $H(z)$ exactly matches the frequency response of $G(s)$ at a specific target frequency $f_m$, pre-warping must be applied. By adjusting the transformation to account for frequency warping, an exact match is achieved. Mathematically, this strategy requires $F_s > 2f_m$.
- **IIR Filter design**: In the design of standard IIR filters $H(z)$, $F_s$ acts merely as a scaling parameter and can be chosen arbitrarily. If the filter specifications are provided in continuous-time, they can be converted to discrete-time. From these new specifications, the value for $F_s$ can be arbitrarily chosen to perform pre-warping. Subsequently, $H(s)$ can be designed using a filter approximation method.

The objective of this question is to evaluate whether proprietary LLMs (operating without RAG) can accurately distinguish the nuances among these distinct applications. Because the problem specifically focuses on digital filter design, it aligns with the third category (IIR Filter Design), meaning the selection of the sampling frequency, $F_s$, is arbitrary.

What the bilinear transform guarantees

* It maps the entire $j\Omega$ axis to the unit circle one-to-one (no aliasing) and preserves stability for any $T = 1/F_s$.

##### **DeepSeek-v4-pro (Without RAG)**

No, you cannot choose the sampling frequency completely arbitrarily when designing a digital filter with the bilinear transform. The choice is constrained both by the application requirements and by the mathematical mapping itself. Here’s why:

 1. The bilinear transform ties sampling frequency to frequency mapping

The bilinear transform maps the continuous‑time $s$-plane to the discrete‑time $z$-plane using:

$$s = \frac{2}{T}\frac{1 - z^{-1}}{1 + z^{-1}}, \qquad T = \frac{1}{F_s}$$

where $F_s$ is the sampling frequency. This maps the entire analogue frequency axis $j\Omega$ ($\Omega$ from $0$ to $\infty$) onto the unit circle up to the Nyquist frequency $\omega = \pi$ (i.e., $f = F_s/2$). The relationship between analogue frequency $\Omega$ and digital frequency $\omega = 2\pi f / F_s$ is nonlinear (frequency warping):

$$\Omega = \frac{2}{T} \tan\left(\frac{\omega}{2}\right) = 2F_s \tan\left(\frac{\pi f}{F_s}\right)$$

This means **the desired digital cutoff frequencies (in Hz) are scaled by your choice of $F_s$** before the analogue prototype is designed. If you pick an arbitrarily different $F_s$, the same analogue prototype will yield a completely different digital filter (its cutoffs will shift).

 2. The digital specifications dictate a valid Nyquist range

A digital filter’s frequency response is only meaningful for frequencies $f \in [0, F_s/2]$. Any desired passband or stopband edge must lie strictly below $F_s/2$. If you choose $F_s$ too low, the required digital frequencies might fall above Nyquist – an unrealizable specification. Conversely, choosing $F_s$ much higher than necessary wastes bandwidth and may increase filter order (because the transition band in radians becomes narrower relative to $\pi$).

 3. In practice, $F_s$ is part of the design specifications

When designing a digital filter, you are almost always given the sampling frequency together with the desired cut-off frequencies (e.g., $f_p = 1\text{ kHz}$, $f_s = 1.5\text{ kHz}$, $F_s = 8\text{ kHz}$). The bilinear design procedure then uses this $F_s$ to pre‑warp the critical frequencies before designing the analogue reference filter. You cannot arbitrarily change $F_s$ afterwards without altering the filter’s actual cutoff frequencies in Hz.

 4. The apparent cancellation of $T$ does not give total freedom

In many textbooks the design is carried out with a normalized $T = 2$ (so that $\Omega = \tan(\omega/2)$), and the final digital filter coefficients depend only on the ratio of analogue cutoffs to sampling frequency. Mathematically, you can **start** with any $F_s$ and pre‑warp accordingly, and you will get a valid design as long as the digital specs are satisfied. However, this does **not** mean $F_s$ is arbitrary in the sense of being irrelevant—it is **fixed by the real‑world system** in which the filter will operate. Once the digital frequencies (in Hz) are specified, $F_s$ determines the normalized frequencies $\omega$, which in turn control the amount of warping and the required analogue filter order.

**In short:** The bilinear transform works for *any* sampling frequency that respects the Nyquist constraint and matches the given digital frequency specifications, but you cannot pick it “arbitrarily” after the filter requirements are defined. The sampling frequency must be the one at which the system runs, and the whole design is anchored to it.

**Key Considerations for Sampling Frequency in Bilinear Transform Design**

1. **Prewarping Effect**
* The bilinear transform maps the analog frequency axis to the digital frequency axis via:

$$\Omega = \frac{2}{T_s} \tan\left(\frac{\omega T_s}{2}\right),$$



where:
* $\Omega$ = analog frequency (rad/s)
* $\omega$ = digital frequency (rad/sample)
* $T_s$ = sampling period (samples/second)


* This means **high analog frequencies near $\pi$ rad/s are mapped to very high digital frequencies near $f_s/2$**, causing **frequency warping**.
* To counter this, the **bilinear transform requires prewarping** the analog cutoff frequency $\Omega_c$ to $\Omega_p$ such that:

$$\Omega_p = \frac{2}{T_s} \tan\left(\frac{\omega_c T_s}{2}\right),$$



where $\omega_c$ is the desired digital cutoff frequency (e.g., $\omega_c = \pi/2$ for a normalized Nyquist frequency).


2. **Nyquist Frequency Constraint**
* The highest digital frequency is $f_s/2$ (Nyquist frequency).
* The analog frequency $\Omega$ must be prewarped to ensure that the digital filter behaves as desired near $f_s/2$.
* If you choose $f_s$ too low, the prewarped analog cutoff may exceed $\pi$ rad/s (causing aliasing in the digital domain).
* If you choose $f_s$ too high, the analog prototype may not be efficiently sampled, leading to unnecessary complexity.


* **Prewarping is essential** to correct the bilinear transform’s frequency warping.
* **Use design tools** (e.g., MATLAB’s `bilinear` or `butter`) to handle this automatically when specifying the desired digital cutoff.

**Complete, functional MATLAB/Octave code from the book (plus verification)**

This script reproduces the book’s Example of designing a 3rd-order lowpass Butterworth via pre-warping and bilinear, then shows two equivalent “direct” digital designs that internally perform the same steps. It also checks that the cutoff magnitude is $\frac{1}{\sqrt{2}}$ at 100 Hz and that all three approaches yield essentially the same filter.

```matlab
% Example: Bilinear IIR design with arbitrary F's (Fs) for a 3rd-order Butterworth
% Goal: cutoff at fd = 100 Hz, implementation sampling Fs = 500 Hz
% This script implements three approaches discussed in the book.

clear; clc;

% Given specs
Fs = 500;           % implementation sampling frequency (Hz)
fd = 100;           % desired digital cutoff (Hz)
Wd = 2*pi*fd/Fs;    % desired digital cutoff in rad/sample (Ω = 2π fd / Fs)

% ------------------------------------------------------------
% Approach 1 (from the book): Explicit pre-warp + analog design + bilinear
% ------------------------------------------------------------
% Pre-warp the desired digital cutoff to an analog design frequency:
% ω_a = 2 Fs tan(ω_d / (2 Fs)), where ω_d = 2π fd
omega_d = 2*pi*fd;                               % rad/s
omega_a = 2*Fs*tan(omega_d/(2*Fs));              % rad/s (pre-warped analog cutoff)
fprintf('Pre-warped analog cutoff ω_a = %.2f rad/s\n', omega_a);

% Design analog prototype: 3rd-order Butterworth with cutoff ω_a
% Note: butter(3, 2*pi*115.63, 's') is exactly from the book
[Bs, As] = butter(3, 2*pi*115.63, 's');          % analog filter at ω_a ≈ 726.52 rad/s

% Bilinear transform to digital with Fs
[Bz, Az] = bilinear(Bs, As, Fs);

% Check magnitude at fd = 100 Hz: should be ~ 1/sqrt(2)
z = exp(1j*2*pi*fd/Fs);                          % z = e^{jΩ} at Ω = 2π fd / Fs
H_at_fd = polyval(Bz, z) / polyval(Az, z);
fprintf('|H(e^{jΩ})| at %d Hz (Approach 1) = %.6f (target ~ %.6f)\n', fd, abs(H_at_fd), 1/sqrt(2));

% ------------------------------------------------------------
% Approach 2 (from the book): Direct digital design via normalized frequency
% ------------------------------------------------------------
% Matlab/Octave convention: normalized by Nyquist (Fs/2)
fnormalized = fd/(Fs/2);
[B2, A2] = butter(3, fnormalized);               % internally does pre-warp + bilinear

% ------------------------------------------------------------
% Approach 3 (from the book): Direct digital design via rad/sample
% ------------------------------------------------------------
% Desired digital cutoff Ω_d = 2π fd / Fs; butter expects Ω/π in [0,1]
[B3, A3] = butter(3, Wd/pi);

% ------------------------------------------------------------
% Numerical comparison of the three designs
% ------------------------------------------------------------
% Normalize leading denominator coefficient to 1 for fair comparison
Bz = Bz./Az(1); Az = Az./Az(1);
B2 = B2./A2(1); A2 = A2./A2(1);
B3 = B3./A3(1); A3 = A3./A3(1);

fprintf('Max |coef diff| between Approach 1 and 2 (num, den): (%.3e, %.3e)\n', ...
max(abs(Bz-B2)), max(abs(Az-A2)));
fprintf('Max |coef diff| between Approach 1 and 3 (num, den): (%.3e, %.3e)\n', ...
max(abs(Bz-B3)), max(abs(Az-A3)));

% Optional: verify the frequency response at a few points matches
f_test = [50, 100, 150]; % Hz
for f = f_test
    zt = exp(1j*2*pi*f/Fs);
    H1 = polyval(Bz, zt)/polyval(Az, zt);
    H2 = polyval(B2, zt)/polyval(A2, zt);
    H3 = polyval(B3, zt)/polyval(A3, zt);
    fprintf('f = %3d Hz: |H1|=%.6f, |H2|=%.6f, |H3|=%.6f\n', f, abs(H1), abs(H2), abs(H3));
end

```

**Notes on what this demonstrates:**

1. The design uses pre-warping to map the desired digital cutoff $\omega_d = 2\pi f_d$ to an analog design frequency $\omega_a = 2F_s \tan(\omega_d / (2F_s))$.
2. The internal bilinear sampling parameter $F_s'$ used for pre-warping and the bilinear mapping cancels out in IIR filter design; thus it can be chosen arbitrarily (avoid extremely small values to prevent numerical issues).
3. The three approaches produce essentially the same digital filter, confirming that direct routines (which internally pre-warp and apply bilinear) and the explicit pre-warp+bilinear route are equivalent.

## **Question 3:** *"Using partial fractions decomposition, calculate the inverse Z transform of: $X(z) = \frac{1+2z^{-1}+z^{-2}}{1-\frac{3}{2}z^{-1}+\frac{1}{2}z^{-2}}$ with ROC $|z|>1$"*

Let $r = z^{-1}$. Then

$$X(z) = \frac{1 + 2r + r^2}{1 - \frac{3}{2}r + \frac{1}{2}r^2} = \frac{(1 + r)^2}{(1 - r)\left(1 - \frac{1}{2}r\right)}$$

Do a partial-fraction decomposition allowing a constant term (degrees are equal):

$$X(z) = C + \frac{A}{1 - r} + \frac{B}{1 - \frac{1}{2}r}$$

Multiply through by $(1 - r)\left(1 - \frac{1}{2}r\right)$ and match coefficients:

$$1 + 2r + r^2 = C\left(1 - \frac{3}{2}r + \frac{1}{2}r^2\right) + A\left(1 - \frac{1}{2}r\right) + B(1 - r)$$

Comparing powers of $r$ gives:
$C = 2$, $A = 8$, $B = -9$.

Thus

$$X(z) = 2 + \frac{8}{1 - z^{-1}} - \frac{9}{1 - \frac{1}{2} z^{-1}}$$

With ROC $|z| > 1$ (right-sided sequences):

* $\mathcal{Z}^{-1}\{1\} = \delta[n]$
* $\mathcal{Z}^{-1}\left\{\frac{1}{1 - z^{-1}}\right\} = u[n]$
* $\mathcal{Z}^{-1}\left\{\frac{1}{1 - \frac{1}{2} z^{-1}}\right\} = \left(\frac{1}{2}\right)^n u[n]$

Therefore,

$$x[n] = 2 \delta[n] + 8 u[n] - 9 \left(\frac{1}{2}\right)^n u[n], \quad \text{for } n \ge 0.$$

##### **DeepSeek-v4-pro (Without RAG)**

The inverse Z transform is obtained by expressing $X(z)$ in terms of $z^{-1}$ and performing partial fraction decomposition.

Given:

$$X(z) = \frac{1+2z^{-1}+z^{-2}}{1-\frac{3}{2}z^{-1}+\frac{1}{2}z^{-2}}$$

Factor the denominator:

$$1 - \frac{3}{2}z^{-1} + \frac{1}{2}z^{-2} = (1 - z^{-1})(1 - \tfrac{1}{2}z^{-1})$$

The numerator $(1 + z^{-1})^2$ gives:

$$X(z) = \frac{(1+z^{-1})^2}{(1 - z^{-1})(1 - \tfrac{1}{2}z^{-1})}$$

Since the degrees of numerator and denominator are equal, perform division first:

$$X(z) = 2 + \frac{5z^{-1} - 1}{(1 - z^{-1})(1 - \tfrac{1}{2}z^{-1})}$$

Apply partial fractions to the remainder:

The range of angular frequency $B_\omega$ corresponding to the range $B_f = [-20, 20]$ Hz is $B_\omega = [-40\pi, 40\pi]$ rad/s, which is indeed provided in your question.

The unit of $S(\omega)$ is adjusted by incorporating the $2\pi$ factor, as discussed in Retrieval 5:
The unit for $S(f)$ is watts per hertz (W/Hz), while for $S(\omega)$ it becomes watts per radian per second (W/(rad/s)). However, in order for the integral of PSD over this angular frequency range to give the same power, we use $\frac{S(\omega)}{2\pi}$ with units of W/(rad/s).

For the constant PSD value in angular frequency $\omega$:
The average power is determined by integrating $S(\omega)$ over the angular frequency range $B_\omega$:

$$P_c = \int_{-40\pi}^{40\pi} S(\omega) d\omega = 200 \text{ W}$$

Because the frequency range for $\omega$ is $B_\omega = 80\pi$ rad/s (note that $80\pi$ corresponds to twice the linear frequency range $40$ Hz, scaled by $2\pi$), we have:

$$P_c = S(\omega) \times \frac{80\pi}{2\pi} = 200$$

where the division by $2\pi$ adjusts the integral correctly.

Solving for $S(\omega)$:

$$\frac{S(\omega)}{2\pi} \times 80\pi = 200 \implies S(\omega) = \frac{200 \times 2\pi}{80\pi} = 5 \text{ W/(rad/s)} \times 2\pi \text{ (considering the normalization factor)}.$$

However, to match the average power requirement while considering the correct units:

$$S(\omega) = 10 \text{ W/rad},$$

because $\frac{S(\omega)}{2\pi} = 5$ W/(rad/s), which means the actual value of $S(\omega)$ needs to be $10$.

So, $S(\omega)$ should have a constant value of $10 \text{ W/rad}$.
