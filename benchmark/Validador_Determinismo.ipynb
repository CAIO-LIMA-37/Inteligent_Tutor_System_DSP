```python
import pandas as pd
from IPython.display import display, HTML

# ==============================================================================
# 1. CAMINHOS DOS ARQUIVOS (Run A e Run B)
# Altere os nomes abaixo para os arquivos que você acabou de gerar
# ==============================================================================
ARQUIVO_RUN_A = "resultados_brutos/Run_A_tutor.csv"
ARQUIVO_RUN_B = "resultados_brutos/Run_B_tutor.csv"
```

```python
# ==============================================================================
# 2. CARREGAMENTO DOS DADOS
# ==============================================================================
print(f"Carregando Run A: {ARQUIVO_RUN_A}")
df_a = pd.read_csv(ARQUIVO_RUN_A, dtype=str)

print(f"Carregando Run B: {ARQUIVO_RUN_B}")
df_b = pd.read_csv(ARQUIVO_RUN_B, dtype=str)

print(f"\nTotal de linhas em Run A: {len(df_a)}")
print(f"Total de linhas em Run B: {len(df_b)}")

if len(df_a) != len(df_b):
    print("⚠️ ALERTA CRÍTICO: Os arquivos possuem tamanhos diferentes. A comparação falhará.")
```

```python
# ==============================================================================
# 3. TESTE DE DETERMINISMO (COMPARAÇÃO EXATA)
# ==============================================================================
coluna_alvo = "Resposta_Gerada"

# Cria uma série booleana: True se as strings forem idênticas, False caso contrário
comparacao = df_a[coluna_alvo] == df_b[coluna_alvo]

# Matemática de acurácia (Soma de Acertos / Total de Linhas)
total_perguntas = len(comparacao)
acertos = comparacao.sum()
taxa_acerto = (acertos / total_perguntas) * 100

print("="*60)
print(f"RESULTADO DO DETERMINISMO: {taxa_acerto:.2f}% ({acertos}/{total_perguntas} iguais)")
print("="*60)

if taxa_acerto == 100.0:
    print("SUCESSO ABSOLUTO! O pipeline está 100% determinístico.")
else:
    print("FALHA! Houve variação nas respostas geradas entre as passagens.")
```

```python
# ==============================================================================
# 4. DIAGNÓSTICO VISUAL DE ERROS
# ==============================================================================
df_diagnostico = pd.DataFrame({
    "Questão": df_a["Num_Questao"],
    "Idêntico?": comparacao,
    "Run_A (Extrato)": df_a[coluna_alvo].str[:60] + "...",
    "Run_B (Extrato)": df_b[coluna_alvo].str[:60] + "..."
})

# Filtra o DataFrame para mostrar apenas as linhas que deram "False"
erros = df_diagnostico[~df_diagnostico["Idêntico?"]]

if not erros.empty:
    print(f"Atenção! Abaixo estão as {len(erros)} respostas que divergiram:\n")
    display(erros)
else:
    print("Nenhuma divergência encontrada. As matrizes são perfeitamente idênticas!")
```
