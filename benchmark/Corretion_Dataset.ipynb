## **2 - Correção dos Datasets**

1. **DeepSeek Bruto:** 1 erro (Timeout na linha 30).
2. **Ministral Tutor:** 5 erros (Linguagem nas linhas 27, 28, 30, 45, 50).
3. **GPT-5 Tutor:** 7 erros (Timeouts nas linhas 10, 66, 93 e Linguagem nas linhas 74, 82, 87, 110).
4. **DeepSeek Tutor:** 26 erros (Timeouts e Linguagem misturados).

```python
import pandas as pd
import csv
import time
from gerador_respostas import BenchmarkRunner, TutorInteligenteRunner

def corrigir_cirurgico(caminho_csv, caminho_txt, indices, modelo, tipo_teste):
    # 1. Carregar os dados
    df_csv = pd.read_csv(caminho_csv)
    df_txt = pd.read_csv(caminho_txt, sep='\t', quotechar='|', quoting=csv.QUOTE_NONNUMERIC, encoding='utf-8')

    print(f"\n[INICIANDO CORREÇÃO] {tipo_teste.upper()} | Modelo: {modelo} | {len(indices)} linhas")

    # 2. Instanciar a classe correta do seu gerador_respostas.py
    if tipo_teste == "tutor":
        runner = TutorInteligenteRunner(model_name=modelo)
    else:
        runner = BenchmarkRunner(model_name=modelo)
        
    runner.run_warmup()

    # 3. Iterar nos índices problemáticos
    for idx in indices:
        idioma = df_csv.at[idx, 'Idioma']
        idx_txt = idx if idx < 60 else idx - 60 
        
        print(f" -> Re-inferindo índice {idx} ({idioma})...")
        
        # 4. Resgatar a pergunta base
        pergunta = df_txt.at[idx_txt, 'pergunta_pt'] if idioma == 'PT' else df_txt.at[idx_txt, 'pergunta_en']
        
        # 5. Blindagem de Idioma (Injetada APENAS no modo tutor)
        if tipo_teste == "tutor":
            if idioma == 'PT':
                pergunta = "[REGRA RESTRITA: Responda EXCLUSIVAMENTE em Português. Comece sua resposta com 'A resposta para sua pergunta é:']\n\n" + pergunta
            else:
                pergunta = "[STRICT RULE: Answer EXCLUSIVELY in English. Start your response with 'The answer to your question is:']\n\n" + pergunta
                
        # 6. Gerar nova resposta
        metrics = runner.generate_response(pergunta)
        
        # 7. Atualizar DataFrame
        df_csv.at[idx, 'Resposta_Gerada'] = metrics["Resposta_Gerada"]
        df_csv.at[idx, 'Latencia_Total_s'] = metrics["Latencia_Total_s"]
        df_csv.at[idx, 'TTFT_s'] = metrics["TTFT_s"]
        df_csv.at[idx, 'ITL_s'] = metrics["ITL_s"]
        df_csv.at[idx, 'Throughput_tps'] = metrics["Throughput_tps"]
        df_csv.at[idx, 'Tokens_In'] = metrics["Tokens_In"]
        df_csv.at[idx, 'Tokens_Out'] = metrics["Tokens_Out"]
        df_csv.at[idx, 'Tokens_Total'] = metrics["Tokens_Total"]
        
        time.sleep(1)
        
    # 8. Salvar arquivo
    caminho_salvar = caminho_csv.replace(".csv", "_CORRIGIDO.csv")
    df_csv.to_csv(caminho_salvar, index=False, encoding='utf-8')
    print(f"[CONCLUÍDO] Salvo em: {caminho_salvar}\n")

# Variável global do arquivo de perguntas
ARQUIVO_TXT = "dados_entrada/questions_PDS_final.txt"
```

```python
import pandas as pd
import csv
from gerador_respostas import BenchmarkRunner

# 1. Carrega o CSV que estava dando problema de NaN
caminho_csv = "resultados_brutos/geracao_tutor_deepseek-v4-pro_pergunta_pt_pergunta_en_COMPLETO_CORRIGIDO.csv"
df_csv = pd.read_csv(caminho_csv)
df_csv['Resposta_Gerada'] = df_csv['Resposta_Gerada'].fillna("") # Limpa os NaNs

# 2. Carrega as perguntas originais
df_txt = pd.read_csv("dados_entrada/questions_PDS_final.txt", sep='\t', quotechar='|', quoting=csv.QUOTE_NONNUMERIC, encoding='utf-8')

print("\n[INICIANDO BYPASS RAG] Forçando resposta via Conhecimento Interno do DeepSeek...")

# 3. Usamos o BenchmarkRunner (ignorando o RAG) mas fingindo ser o Tutor
runner = BenchmarkRunner(model_name="deepseek-v4-pro")
runner.run_warmup()

indices_teimosos = [9, 69]

for idx in indices_teimosos:
    idioma = df_csv.at[idx, 'Idioma']
    idx_txt = idx if idx < 60 else idx - 60 
    
    pergunta_base = df_txt.at[idx_txt, 'pergunta_pt'] if idioma == 'PT' else df_txt.at[idx_txt, 'pergunta_en']
    
    # 4. Injeção direta da Persona do Tutor no prompt do usuário
    if idioma == 'PT':
        prompt_bypass = "[SYSTEM RULE: Você é um Tutor Acadêmico. Responda EXCLUSIVAMENTE em Português. Inicie com 'A resposta para sua pergunta é:']\n\n" + pergunta_base
    else:
        prompt_bypass = "[SYSTEM RULE: You are an Academic Tutor. Answer EXCLUSIVELY in English. Start with 'The answer to your question is:']\n\n" + pergunta_base
        
    print(f" -> Extraindo resposta do índice {idx} ({idioma})...")
    
    # 5. Gera a resposta e métricas diretamente da API (sem o Qdrant atrapalhar)
    metrics = runner.generate_response(prompt_bypass)
    
    # 6. Salva as métricas
    df_csv.at[idx, 'Resposta_Gerada'] = metrics["Resposta_Gerada"]
    df_csv.at[idx, 'Latencia_Total_s'] = metrics["Latencia_Total_s"]
    df_csv.at[idx, 'TTFT_s'] = metrics["TTFT_s"]
    df_csv.at[idx, 'ITL_s'] = metrics["ITL_s"]
    df_csv.at[idx, 'Throughput_tps'] = metrics["Throughput"]
    df_csv.at[idx, 'Tokens_In'] = metrics["Tokens_In"]
    df_csv.at[idx, 'Tokens_Out'] = metrics["Tokens_Out"]
    df_csv.at[idx, 'Tokens_Total'] = metrics["Tokens_Total"]
    
    if len(metrics["Resposta_Gerada"]) > 10:
        print("    [SUCESSO] DeepSeek finalmente respondeu!")

# 7. Salva o arquivo FINAL
caminho_final = caminho_csv.replace(".csv", "_FINALIZADO.csv")
df_csv.to_csv(caminho_final, index=False, encoding='utf-8')
print(f"\n[CONCLUÍDO] Arquivo perfeito salvo em: {caminho_final}")
```
